```python
# Leave this code block unchanged
from pathlib import Path
import pandas as pd

project_root = Path().resolve().parent  # if notebook is in /notebooks folder

csv_path = project_root/"data"/"raw"/"healthcare-dataset-stroke-data.csv"

df_raw = pd.read_csv(csv_path)

print(df_raw.info())
```

```python
"""
Cram√©r's V between age_clinical_bin and stroke: 0.2670"
"""
import seaborn as sns
sns.catplot(data=df_raw, x="stroke", y="age", kind="strip", alpha = .5)
```

```python
# age vs ever_married vs work_type (high Theil's U from Sweetviz)
sns.boxplot(data=df_raw, x="work_type", y="age", hue="ever_married")
```
